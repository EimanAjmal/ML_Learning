```python
import pandas as pd
import matplotlib.pyplot as plt

# first load the data
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic = pd.read_csv(url)

# then analyze the data
survival_by_gender = titanic.groupby("Sex")["Survived"].mean() * 100

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
survival_by_gender.plot(kind="bar", color=["blue", "pink"])
plt.title("Survival Rate by Gender")
plt.ylabel("Survival %")
plt.xticks(rotation=0)

plt.subplot(1, 2, 2)
titanic["Pclass"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Passenger Class Distribution")

plt.tight_layout()
plt.show()

print(f"Female Survival: {survival_by_gender['female']:.1f}%")
print(f"Male Survival  : {survival_by_gender['male']:.1f}%")
```
