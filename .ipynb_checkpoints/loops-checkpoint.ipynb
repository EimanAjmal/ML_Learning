```python
# Function — Repeatition
def greet(name):
    print(f"Assalam o Alaikum, {name}!")
    print(f"Welcome to ML Learning!")

# call function
greet("Eiman")
greet("Sara")
greet("Ali")
```

```python

def calculate_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "Fail"


def student_info(name, marks, subject):
    grade = calculate_grade(marks)
    print(f"Student : {name}")
    print(f"Subject : {subject}")
    print(f"Marks   : {marks}")
    print(f"Grade   : {grade}")
    print("-" * 25)


student_info("Eiman", 92, "Machine Learning")
student_info("Sara", 78, "Data Science")
student_info("Ali", 55, "Python")
```
