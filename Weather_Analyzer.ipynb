```python
# Project 2 — Weather Analyzer
import numpy as np
import matplotlib.pyplot as plt

# Karachi weekly temperature
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temp = [32, 35, 33, 36, 34, 38, 37]
humidity = [60, 55, 65, 58, 70, 52, 48]

# Analysis
print("=" * 40)
print("   KARACHI WEATHER REPORT CARD")
print("=" * 40)
print(f"Average Temp    : {np.mean(temp):.1f}°C")
print(f"Hottest Day     : {days[np.argmax(temp)]} ({max(temp)}°C)")
print(f"Coolest Day     : {days[np.argmin(temp)]} ({min(temp)}°C)")
print(f"Average Humidity: {np.mean(humidity):.1f}%")
print("=" * 40)

# Graph
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Temperature graph
ax1.plot(days, temp, color="red", marker="o", linewidth=2)
ax1.fill_between(days, temp, alpha=0.3, color="red")
ax1.set_title("Temperature This Week")
ax1.set_xlabel("Days")
ax1.set_ylabel("Temperature (°C)")
ax1.grid(True)

# Humidity graph
ax2.bar(days, humidity, color="green", alpha=0.7)
ax2.set_title("Humidity This Week")
ax2.set_xlabel("Days")
ax2.set_ylabel("Humidity (%)")
ax2.grid(True)

plt.tight_layout()
plt.show()
```
